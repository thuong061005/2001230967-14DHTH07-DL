```python
#bai12
import tensorflow as tf


def create_random_tensors():

    x = tf.random.normal(shape=(3, 2, 2))
    y = tf.random.normal(shape=(3, 2, 2))
    return x, y


@tf.function
def math_operations_graph(x, y):

    addition = tf.add(x, y)
    subtraction = tf.subtract(x, y)
    multiplication = tf.multiply(x, y)
    division = tf.divide(x, y)

    return addition, subtraction, multiplication, division


tensor_x, tensor_y = create_random_tensors()


res_add, res_sub, res_mul, res_div = math_operations_graph(tensor_x, tensor_y)


print("Tensor X ngẫu nhiên:\n", tensor_x.numpy())
print("\nKết quả phép cộng (Graph Execution):\n", res_add.numpy())
```

```python
#bai13
import tensorflow as tf


class AdditionModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)


    @tf.function
    def __call__(self, x, y):

        return tf.add(x, y)

sum_model = AdditionModel(name="sum_model")


x = tf.constant([10, 20], dtype=tf.int32)
y = tf.constant([5, 15], dtype=tf.int32)


result = sum_model(x, y)

print("Tensor x:", x.numpy())
print("Tensor y:", y.numpy())
print("Kết quả x + y từ model:", result.numpy())
```

```python
#bai14
import tensorflow as tf

class MathModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)


    @tf.function
    def __call__(self, x, y):

        diff = tf.subtract(x, y)


        prod = tf.multiply(x, y)

        return diff, prod


my_math_model = MathModel(name="math_operations")

x = tf.constant([10, 20, 30], dtype=tf.float32)
y = tf.constant([2, 4, 6], dtype=tf.float32)

difference, product = my_math_model(x, y)

print("Tensor x:", x.numpy())
print("Tensor y:", y.numpy())
print("Hiệu (x - y):", difference.numpy())
print("Tích (x * y):", product.numpy())
```

```python
#bai15
import tensorflow as tf


class RectanglePerimeterModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)


    @tf.function
    def __call__(self, a, b):

        sum_sides = tf.add(a, b)

        perimeter = tf.multiply(sum_sides, 2.0)

        return perimeter

model = RectanglePerimeterModel(name="perimeter_calculator")

a = tf.constant(10.0, dtype=tf.float32)
b = tf.constant(5.0, dtype=tf.float32)


result = model(a, b)

print(f"Chiều dài a: {a.numpy()}")
print(f"Chiều rộng b: {b.numpy()}")
print(f"Chu vi hình chữ nhật P: {result.numpy()}")
```

```python
#bai16
import tensorflow as tf


class BoxAreaModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)


    @tf.function
    def __call__(self, a, b, c):

        sum_ab = tf.add(a, b) #
        s_xung_quanh = tf.multiply(tf.multiply(sum_ab, c), 2.0) #


        s_day = tf.multiply(tf.multiply(a, b), 2.0)
        s_toan_phan = tf.add(s_day, s_xung_quanh)

        return s_xung_quanh, s_toan_phan


model = BoxAreaModel(name="box_area_calculator")

a = tf.constant(5.0, dtype=tf.float32)
b = tf.constant(3.0, dtype=tf.float32)
c = tf.constant(4.0, dtype=tf.float32)


sxq, stp = model(a, b, c)

print(f"Kích thước hộp: dài={a.numpy()}, rộng={b.numpy()}, cao={c.numpy()}")
print(f"Diện tích xung quanh: {sxq.numpy()}")
print(f"Diện tích toàn phần: {stp.numpy()}")
```

```python
#bai17
import tensorflow as tf

class ParallelSidesModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)

    @tf.function
    def __call__(self, AD, BC, AB, CD):

        cond1 = tf.equal(AD, BC)

        cond2 = tf.equal(AB, CD)

        is_parallel = tf.logical_and(cond1, cond2)
        return is_parallel


model_17 = ParallelSidesModel(name="check_parallel")
result = model_17(tf.constant(5), tf.constant(5), tf.constant(3), tf.constant(3))
print("Kết quả chứng minh AD // BC:", result.numpy())
```

```python
#bai18
import tensorflow as tf

class EqualBasesModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)

    @tf.function
    def __call__(self, AD, BC, is_AD_parallel_BC):

        is_equal_sides = tf.equal(AD, BC)

        is_equal_bases = tf.logical_and(is_equal_sides, is_AD_parallel_BC)
        return is_equal_bases


model_18 = EqualBasesModel(name="check_equal_bases")

is_parallel = tf.constant(True)
result = model_18(tf.constant(4), tf.constant(4), is_parallel)
print("Kết quả chứng minh AB = CD:", result.numpy())
```

```python
#bai19
import tensorflow as tf

class TriangleMidpointModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)

    @tf.function
    def __call__(self, AM, MB, is_MN_parallel_BC):
        """
        Chứng minh NA = NC dựa trên:
        1. M là trung điểm của AB (AM = MB)
        2. MN song song với BC
        """

        is_M_midpoint = tf.equal(AM, MB)

        can_conclude_NA_equal_NC = tf.logical_and(is_M_midpoint, is_MN_parallel_BC)

        return can_conclude_NA_equal_NC


model = TriangleMidpointModel(name="midpoint_proof_model")

AM_val = tf.constant(5.0)
MB_val = tf.constant(5.0)
parallel_cond = tf.constant(True)

result = model(AM_val, MB_val, parallel_cond)

print(f"Điều kiện AM = MB: {tf.equal(AM_val, MB_val).numpy()}")
print(f"Điều kiện MN // BC: {parallel_cond.numpy()}")
print(f"Kết luận NA = NC có đúng không? -> {result.numpy()}")
```
